```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def A(Jz, B, beta, m):
    return -(1/beta)*np.log(2*np.cosh(beta*(Jz*m + B))) + Jz*m**2/2
```

```python
def plotter(Jz, B, beta):
    
    m = np.linspace(-2,2,1000)
    y = -(1/beta)*np.log(2*np.cosh(beta*(Jz*m + B))) + Jz*m**2/2
    plt.plot(m,y)
    plt.xlim([-1.5,1.5])
    plt.xlabel('m')
    plt.ylabel(r'$a_{mf}$')
    
    # Find min point
    min_point = (y.min(), m[y.argmin()])
    
    
    plt.title(f'min A: {min_point[0]:.0f}, m_min: {min_point[1]:.2f}')
```

```python
plotter(Jz = 10, B = 10, beta = 0.00001)
print('Large field, very high temp')
print('As we expect, entropy wins out')
```

```python
plotter(Jz = 10, B = 10, beta = 0.01)
print('Large field, still high but lower temp')
print('Minimum goes off m = 0')
```

```python
plotter(Jz = 10, B = 10, beta = 0.1)
print('Large field, even lower temp')
print('Temp low enough now so that magnetic field wins out against entropy')
print('Everything in line with expectations so far')
```
